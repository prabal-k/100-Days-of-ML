```python
from selenium import webdriver  #Imports the webdriver module from Selenium, which is used to control the browser.
from selenium.webdriver.common.by import By  #Provides methods to locate elements on a webpage (e.g., by ID, name, class name, etc.).
from selenium.webdriver.common.keys import Keys  #Contains keyboard keys (e.g., ENTER, ESCAPE, etc.) to simulate typing actions.
import time  #Python's built-in time module is used for delays (e.g., time.sleep() to wait for a webpage to load).
```

```python
driver = webdriver.Chrome()
driver.get('https://wiki.erepublik.com/index.php/List_of_cities_and_regions')
```

## To get the name of columns of table(i.e table heading)

```python
titles = []
## XPATH to table head(i.e 'thead')
table_head = driver.find_element(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table/thead')
table_rows = table_head.find_elements(By.TAG_NAME,'th')
for th in table_rows:
    titles.append(th.text)
```

```python
titles
```

## To get the content of the table (i.e table body)

```python
## There is a single table body so xpath to table 
table_body = driver.find_element(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table/tbody')
# Find all rows in the table body
table_rows = table_body.find_elements(By.TAG_NAME, 'tr')
data=[]
# Iterate over each row
for table_row in table_rows:
    # Find all cells in the current row
    cells = table_row.find_elements(By.TAG_NAME, 'td')
    row = [c.text for c in cells]
    data.append(row)
   
    
data
```

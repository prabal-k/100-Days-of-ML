## IQR technique is used to remove outlier when the data is skewed

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import scipy.stats as stats     #To draw the QQ plot    
```

```python
df = pd.read_csv('Data-sets\\placement(1).csv')
df
```

## we used z-scored for 'cgpa' in previous notebook
## Now we will use iqr for 'placement_exam_marks'

```python
sns.histplot(data=df,x='placement_exam_marks',kde=True)
plt.title('Distribution of placement_exam_marks')
plt.show()

# To draw the QQ plot for the placement_exam_marks columnto identify the distribution 
plt.figure(figsize=(8,8))
stats.probplot(df['placement_exam_marks'],dist='norm',plot=plt)
plt.title('QQ plot placement_exam_marks')
plt.show()

```

## Here 'placement_exam_makrs' are skewed

```python
print("Skewness of placement_exam_marks: ",df['placement_exam_marks'].skew())
# It is right skewed
```

# Box-plot to visualize outlier in placement_exam_marks

```python
sns.boxplot(data=df,x='placement_exam_marks')
```

# Uisng IQR to find upper and lower bound

```python
df['placement_exam_marks'].describe()
```

```python
q1 = df['placement_exam_marks'].quantile(0.25)  # Q1= 25th percentile
q3 = df['placement_exam_marks'].quantile(0.75) # Q3= 75th percentile
print("q1 of placement_exam_marks: ",q1)
print("q3 of placement_exam_marks: ",q3)
```

```python
iqr = q3-q1
lb = q1-1.5*iqr
ub = q3+1.5*iqr
print("IQR: ",iqr)
print("lower bound ",lb)
print("upper bound: ",ub)
```

## The outlier present rows are

```python
df[(df['placement_exam_marks']<lb) | (df['placement_exam_marks']>ub)]
```

# Trimming

```python
new_df = df[(df['placement_exam_marks'] > lb) & (df['placement_exam_marks'] < ub)]
new_df
```

```python
new_df.shape
```

```python
# Comparing the distribution and box plot after and before the outlier detection and removal

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```

# Trimming

```python
# np.where(x,y,z)
# where x=condition
# y = statement when condition is true
# z = statement when condition is false
```

```python
new_df2 = df.copy()
new_df2['placement_exam_marks']=np.where(
    df['placement_exam_marks'] > ub,   #condition when value is greater than upper bound
    ub,                # then replace it by upper bound
    np.where(                   # if above condition is false then another condition to check if value is smaller than lower bound
        df['placement_exam_marks'] < lb,
        lb,            #replace by lower bound
        df['placement_exam_marks']              # incase of above both condition are not satisfied(i.e non-outlier) keep the original value
    )
)

new_df2
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df2['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df2['placement_exam_marks'])

plt.show()
```

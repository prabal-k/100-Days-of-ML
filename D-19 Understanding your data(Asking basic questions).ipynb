```python
titanic.info()
```

```python
cat_col = [c for c in titanic.columns if titanic[c].dtype=='object']
num_col = [c for c in titanic.columns if titanic[c].dtype!='object']
print('Categorical columns: ',cat_col)
print('Numerical columns: ',num_col)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Ignoring age and gender columns for now 
x=df.iloc[:,2:4]
y=df.iloc[:,4:]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# We create an object of Ordinal Encoder
# In categories paramater there comes the values of 2 columns that we want to transform
# In first column(review) we want Poor ro get the lowest value then slightly higher to Average and highest to Good so we keep in this order(1st poor then Average then good)
# If we do not specify the categories paramater than randomly among good,average and poor will get the highest value/order
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
#After we perform train test split we fit the model on training data and transform both training ad testing data becaue:
# --->Fitting: Learn patterns from training data only.
# --->Transforming: Apply learned patterns to both training and testing data for consistency and fair evaluation.


# 2. Reasons for Fitting Only on Training Data:

#     Data Leakage: Fitting the model on both the training and testing data can introduce data leakage. This occurs when information from the testing set unintentionally influences the model's learning process, leading to overfitting and poor generalization.
#     Evaluation Bias: Evaluating the model's performance on data it has already seen (the testing set) can provide a biased estimate. The model might perform better than expected on unseen data.
```

```python
oe.categories_
```

```python
#Now we transform both training and test data
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

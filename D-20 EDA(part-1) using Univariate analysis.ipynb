```python
#Univariate analysis-> analysis of single varibale/column at a time
#bivariate analysis-> analysis of 2 variable/columns at a time
#multivariate analysis-> analysis of more than 2 variable/column at a time
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

# 1. Univariate analysis of categotical data

#### for univariate analysis of categorical data we mostly use: count plot (countplot()) and piechart

```python
cat_col = [c for c in df.columns if df[c].dtype=='object']
print("categorical columns :",cat_col)
```

# a.count plot

```python
# to determive the count of survived people
sns.countplot(data=df,x='survived')
```

```python
# death = 549 
# survived = 342
```

```python
# to count the number of people in pclass
sns.countplot(data=df,x='pclass')
```

```python
df[df['survived']==0].groupby('sex').agg({'survived':'count'})
```

# b. Pie chart

```python
# To show the percentage of survived people
val = df['survived'].value_counts()
val
plt.pie(x=val,autopct='%.2f',labels=['DIED','SURVIVED'])
plt.show()
```

```python
# To show the percentage of survived people
df['survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# To show the percentage of pclass(passenger class)

df['pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Univariate Analysis of Numerical data

#### IN univariate analysis of numerical data we can plot:histogram ,kde, boxplot,rugplot ,violin plot etc
#### here we also calculate the minimux value,max value ,mean 

```python
num_col = [c for c in df.columns if df[c].dtype !='object']
print("numerical columns: ",num_col)
```

# A. Histogram

```python
# distribution of age using histogram
sns.displot(data=df,x='age',rug=True,bins=10,color='black')
```

```python
#Knowledge: Maximum number of people where from age between 20 to 40
```

# B. Kde

```python
sns.displot(df['age'],kind='kde')
```

```python
sns.distplot(df['age'])
```

```python
#Knowledge: Kde plot tells about the probability of having of number (in kde in y-axis there is probability and x-axis is frequency)
# for eg : If we randomly choose a passenger then the probability that his age would be 40 is 1.5 percent
```

# C. Box plot

### box plot gives the 5 number summary: min,q1,median,q3,max

```python
sns.catplot(data=df,x='fare',kind='box')
```

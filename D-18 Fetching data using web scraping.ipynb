# Step -1 Get the Html


```python
import requests
from bs4 import BeautifulSoup
```

```python
# THIS HEADER is used to act like the requests are comming tgrough a web browser 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
url = "https://www.kistcollege.edu.np"  # Website that we want to scrap
url
```

```python
r = requests.get(url,headers=headers)
print(r)
# Response is 200 means The server has successfully processed the request, and the requested resource (in this case, the webpage) has 
# been returned in the response. This is the standard response for successful HTTP requests
```

```python
r.status_code
```

###### What's Happening: The requests.get(url) function is being called to send an HTTP GET request to the specified URL ("https://www.kistcollege.edu.np").

What It Does: This function attempts to retrieve the content of the webpage located at the given URL. It returns a Response object that contains various details about the response, such as the content of the page, status code, headers, etc.###### 
Variable r: The variable r stores the Response object returned by the requests.get call.

```python
htmlcontent = r.content
htmlcontent
# reterives the entire HTML content of the webpage that was fetched.
```

# Step-2 Parse the HTML

```python
soup = BeautifulSoup(htmlcontent , 'html.parser',)
soup
```

```python
# prettify , pretifies the html code (Make the proper indentation ,format the code etc)
soup.prettify
```

# Step-3 HTML Tree Traversal

```python
# Commonly used types of object:
# 1. Tag 
# 2. NavigableString
# 3. BeautifulSoup
# 4. Comment
```

```python
print(type(soup))  # 3. BeautifulSoup

print(type(soup.title)) # 1. Tag 

print(type(soup.title.string))# 2. NavigableString
```

```python
print(soup.title) # returns title of the html page
```

```python
print(soup.title.string)  # returns string title of the html page
```

```python
# Displays the name of all tags prenet in the html
for tag in soup.find_all():
    print(tag.name)
```

```python
# get all the paragraphs (not only paragraphs it can be anchor tags etc) from the web page
paragraphs=soup.find_all('p')
print(paragraphs)
```

```python
# get all the anchor tags from the web page
paragraphs=soup.find_all('a')
print(paragraphs)
```

```python
soup.find('p') # Returns the first paragraph tag of the html page
```

```python
soup.find('p')['class'] # Returns all classes of the first paragraph tag of the html page
```

```python
# returns the first anchor tag of the html page
soup.find_all('p',attrs={'class':"text-md font-medium tracking-wide text-white text-center"})
```

# Opening csv file from an URL

```python
import requests   #This line imports the requests library, which is a Python library used for making HTTP requests.
from io import StringIO  # imports the StringIO class from the io module. StringIO is a class that creates a file-like object in memory that can be used as input or output.

url = "https://raw.githubusercontent.com/prabal-k/Python-Pandas-numpy-matplotlib-/main/Data%20sets%20and%20notes/Data%20sets/company1.csv"  # URL of a raw CSV file hosted on GitHub (not the link of the github webpage )
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"} #This line creates a dictionary called headers, which contains a key-value pair. The key is "User-Agent", and the value is a user-agent string. User-agent headers are used to identify the client making the HTTP request. In this case, it's pretending to be Mozilla Firefox on macOS.
req = requests.get(url, headers=headers) # sends an HTTP GET request to the URL specified in the url variable. It uses the requests.get() function from the requests library. The headers parameter is used to pass the headers dictionary to the request.
data = StringIO(req.text) #This line creates a StringIO object called data and initializes it with the text content of the HTTP response (req.text). This is possible because req.text contains the content of the response in string format.

df1=pd.read_csv(data) #to read the CSV data from the data variable into a DataFrame called df1. The read_csv() function is used to read CSV files into DataFrame objects.
df1
```
